```python

def count_overdue_complaints(df):
    count = 0
    overdue_complaints = []

    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        complaint_id = row["id"]
        status = row["status"]
        color = row["color"]
        due_date = row["due_date"]
        closed_date = row["Closed_date"]

        # Check if the complaint is not marked as red and has a due date
        if (pd.isnull(color) or color.lower() not in ["red", "dark red"]) and pd.notnull(due_date):
            # Check if the complaint is open or closed and went overdue
            if status.lower() in ["open", "draf", "abc"]:
                if pd.notnull(closed_date):
                    if pd.to_datetime(closed_date) > pd.to_datetime(due_date):
                        count += 1
                        overdue_complaints.append(index)
                else:
                    if current_date > pd.to_datetime(due_date):
                        count += 1
                        overdue_complaints.append(index)

    # Create a new dataframe containing only the overdue complaints
    overdue_df = df.loc[overdue_complaints]

    return count, overdue_df

# Set the current date
current_date = pd.to_datetime("2023-05-23")

# Call the function and get the count and overdue dataframe
overdue_count, overdue_df = count_overdue_complaints(complaints_df)

print(f"Number of complaints that were never marked as red but went overdue: {overdue_count}")
print("Overdue Complaints:")
print(overdue_df)
```
